# Problem formulation

$$
\min_{\Omega\subset\mathcal{D}} J(\Omega) \quad \text{subject to} \quad C(\Omega)=\vec{1}
$$

### Notations
$$
	e(w)=\frac{1}{2}(Dw+{Dw}^{\top}),\quad \sigma(w)=2\mu\,e(w)+\lambda(\nabla\cdot w)I
$$

Velocity equation
$$
\mathcal{B} (\theta, \xi)= - \int_{\mathcal{D}} S_0\cdot\xi + S_1 : D\xi \quad\text{for all}\; \xi \in \mathbb{H}(\Omega)
$$

Transport equation

# Compliance minimization with volume constraint
We use
$A_{\Omega} = \chi_{\Omega} + {10}^{-3}\chi_{\mathcal{D}\backslash\Omega}$.

* PDE: find $u \in  H^1_{\Gamma_0}(\mathcal{D}, \mathbb{R}^d)$ such that
	$$
		\int_{\mathcal{D}} A_{\Omega}\sigma(u):e(v) - \int_{\partial \mathcal{D}} g \cdot v
		\qquad
		\forall \, v \in  H^1_{\Gamma_0}(\mathcal{D}, \mathbb{R}^d)
	$$

* Cost functional and its derivative components:
	$$
		J(\Omega) = \int_{\mathcal{D}} A_{\Omega}\sigma(u):e(u)
	$$
	$$
		\begin{aligned}
			S_0 &= \vec{0} \\
			S_1 &= A_{\Omega} (2{Du}^{\top}\sigma(u)- \sigma(u):e(v))I
		\end{aligned}
	$$

* Constraint functional and its derivative components:
	$$
		C(\Omega) = \frac{1}{V} \int_{\mathcal{D}} \chi_{\Omega}
	$$
	$$
		\begin{aligned}
			S_0 &= \vec{0} \\
			S_1 &= \frac{1}{V} \chi_{\Omega} I
		\end{aligned}
	$$

* Bilinear form:
	$$
		B(\theta, \xi) =  {10}^{-1} \int_{\mathcal{D}} \theta \cdot \xi + 
			\int_{\mathcal{D}} D\theta : D\xi +
			{10}^4 \int_{\partial \mathcal{D}}  (\theta \cdot n) (\xi \cdot n)
	$$
	with $\theta, \xi \in H^1(\mathcal{D}, \mathbb{R}^d)$

# Elasticity inverse problem

We use
$A_{\Omega} = 10\chi_{\Omega} + \chi_{\mathcal{D}\backslash\Omega}$.

* `f`-problem: find $u \in  H^1_{\Gamma_0}(\mathcal{D}, \mathbb{R}^d)$ such that
	$$
		\int_{\mathcal{D}} A_{\Omega}\sigma(u):e(w) =
		\int_{\partial \mathcal{D}} f \cdot w
		\qquad
		\forall \, w \in  H^1_{\Gamma_0}(\mathcal{D}, \mathbb{R}^d)
	$$

* `g`-problem: find $v \in  H^1_0(\mathcal{D}, \mathbb{R}^d)$ such that
	$$
		\int_{\mathcal{D}} A_{\Omega}\sigma(v):e(w) =
		- \int_{\mathcal{D}} A_{\Omega}\sigma(g):e(w)
		\qquad
		\forall \, w \in  H^1_0(\mathcal{D}, \mathbb{R}^d)
	$$

* Adjoint of `f`-problem: find $p \in  H^1_{\Gamma_0}(\mathcal{D}, \mathbb{R}^d)$ such that
	$$
		\int_{\mathcal{D}} A_{\Omega}\sigma(p):e(r) =
		-\alpha \int_{\mathcal{D}} (u-v-g) \cdot r
		-\beta \int_{\Gamma_1} (u - g) \cdot r
		\qquad
		\forall \, r \in  H^1_{\Gamma_0}(\mathcal{D}, \mathbb{R}^d)
	$$

* Adjoint of `g`-problem: find $q \in  H^1_0(\mathcal{D}, \mathbb{R}^d)$ such that
	$$
		\int_{\mathcal{D}} A_{\Omega}\sigma(q):e(r) =
		\alpha \int_{\mathcal{D}} (u-v-g) \cdot r
		\qquad
		\forall \, r \in  H^1_0(\mathcal{D}, \mathbb{R}^d)
	$$

* Cost functional and its derivative components:
	$$
		J(\Omega) = \frac{\alpha}{2}\int_{\mathcal{D}}\left|u-v-g\right|^{2} +
		\frac{\beta}{2}\int_{\Gamma_{1}}\left|u-g\right|^{2}
	$$
	$$
		\begin{aligned}
			S_0 &= -\alpha{Dg}^{\top}(u-v-g) +
			A_{\Omega}\sum_{i,j=1}^{d}{\sigma (q)}_{i,j}\nabla(e(g)_{i,j}) \\
			S_1 &= \left(\frac{\alpha}{2}{|u-v-g|}^2 + A_{\Omega}\sigma(u):e(p) + A_{\Omega}\sigma(v+g):e(q)\right)I\\
			{} & 
			\quad + A_{\Omega}\left({Du}^{\top}\sigma(p) + {Dp}^{\top}\sigma(u) + {Dv}^{\top}\sigma(q) + {Dq}^{\top}\sigma(v+g)\right)
		\end{aligned}
	$$
* Bilinear form:
	$$
		B(\theta, \xi) =  \int_{\mathcal{D}} \theta \cdot \xi + 
							\int_{\mathcal{D}} D\theta : D\xi
	$$
	with $\theta, \xi \in H^1_{0}(\mathcal{D}, \mathbb{R}^d)$

# Heat conduction problem:

We use
$A_{\Omega} = \chi_{\Omega} + {10}^{-2}\chi_{\mathcal{D}\backslash\Omega}$.

* PDE: find $u \in  H^1_{\Gamma_0}(\mathcal{D})$ such that
	$$
		\int_{\mathcal{D}} A_{\Omega} \nabla u \cdot \nabla v = 
		\int_{\partial \mathcal{D}} f \cdot v
		\qquad
		\forall \, v \in  H^1_{\Gamma_0}(\mathcal{D})
	$$

* Cost functional and its derivative components:
	$$
		J(\Omega) = \int_{\mathcal{D}} f\,u
	$$
	$$
		\begin{aligned}
			S_0 &= 2u \nabla f \\
			S_1 &= (2uf-A_{\Omega}{|\nabla u|}^2)I+2A_{\Omega}\nabla u\otimes\nabla u
		\end{aligned}
	$$

* Constraint functional and its derivative components:
	$$
		C(\Omega) = \frac{1}{V} \int_{\mathcal{D}} \chi_{\Omega}
	$$
	$$
		\begin{aligned}
			S_0 &= \vec{0} \\
			S_1 &= \frac{1}{V} \chi_{\Omega} I
		\end{aligned}
	$$

* Bilinear form:
	$$
		B(\theta, \xi) =  {10}^{6}\int_{\mathcal{D}} \chi_{\Omega_0}\theta \cdot \xi + 
			\int_{\mathcal{D}} D\theta : D\xi
	$$
	with $\theta, \xi \in H^1_0(\mathcal{D}, \mathbb{R}^d)$ and $\Omega_0$ a neighbourhood of $\Gamma_0$.

**Notes:**

The shape derivatives must have the form:
$$
dJ(\Omega) \theta = \int_{\mathcal{D}} S_0^J \cdot \theta + S_1^J : D\theta 
$$
$$
dC(\Omega) \theta = \int_{\mathcal{D}} S_0^C \cdot \theta + S_1^C : D\theta 
$$


Velocity problem:
Find $\theta \in H$ such that
$$
B(\theta, \xi) = -dJ(\Omega) \xi - \lambda \,dC(\Omega) \xi \qquad \forall\,\xi\in H 
$$
where $\lambda$ is a multiplier and $H = H^1(\Omega, \mathbb{R}^d)$ or $H = H_0^1(\Omega, \mathbb{R}^d)$. The bilinear form is given.




* The `Level` class implements methods to solve two versions of the level set equation. The first method approximates the solution of the classic level set equation with weakly imposed boundary condition. The second method attempts to approximate the level set function by considering the level set equation with a time-varying velocity field. No boundary conditions are applied in the latter case.

* We might consider to solve the transport equation with a diffusive term. See Schmalz1983.pdf p. 15.
